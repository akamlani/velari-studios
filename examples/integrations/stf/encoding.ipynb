```python
import numpy as np
import pandas as pd
# moduel packages
from   velari.integrations.stf.encoding    import Encoder
from   velari.integrations.sklearn.scoring import scorer_cosine_sim
```

```python
sentences = [
    'The cat sat on a mushy mat',
    'The cat stretched and laid down on a furry mat'
]
df_corpus = pd.DataFrame(sentences, columns=['text'])
```

```python
# creat an 'emb' column with the sentence embeddings
encoder   = Encoder()
df_corpus = df_corpus.assign(
    emb=df_corpus['text'].apply(lambda x: encoder.encode(x))
)
```

```python
scorer_cosine_sim(df_corpus.loc[0, 'emb'], df_corpus.loc[1, 'emb'])
```
